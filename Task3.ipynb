# Dropping Irrelevant Column

```python
col= ['Postal Code']
data1=Data.drop(columns=col,axis=1)
```

# Checking statistical relation among rows and columns

```python
# Correlation between variables
data1.corr()
```

```python
data1.cov() #Covariance
```

```python
data1.head() # Load first five rows
```

```python
plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category',data=data1)
plt.title('Category vs Sub Category')
plt.xlabel('Sub-Category')
plt.ylabel('Category')
plt.xticks(rotation=45)
plt.show()
```

```python
data1.hist(bins=50,figsize=(20,15))
plt.show();
```

```python
# Count the total repeatable states
data1['State'].value_counts()
```

```python
plt.figure(figsize=(15,15))
sns.countplot(x=data1['State'])
plt.xticks(rotation=90)
plt.title("STATE")
plt.show()
```

```python
figsize=(15,10)
sns.pairplot(data1,hue='Sub-Category')
plt.show
```

From the above graph we can say that our data is not normally distributed and it has some amount of outliers too.
